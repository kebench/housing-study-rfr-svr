## Combining all the datasets into one

After doing EDA on each of the dataset and ensuring that the common column (Year) is distinct, the researcher merged the datasets into one using the reduce function since multiple dataframes cannot be merged using a single merge method.

The Year column was also typecasted to an Object since the researcher has no intention of using the Year with numerical operations.

```python
from functools import reduce

dataframes = [avg_housing_df, GDP_df, const_cost_idx_df, loans_df, inflation_df]
# Retrieved from https://realpython.com/python-reduce-function/
# This is the same as javascript's reduce function which executes a function 
# on each eleements of the array and the previous result from the function and returns a single output.
# Very useful for performing accumulated actions such as adding all the elements of the array.
# No need to explicitly use loops.
housing_df = reduce(lambda left, right: pd.merge(left, right, on="Year", how="outer"), dataframes)
housing_df["Year"] = housing_df["Year"].astype("object")
```

The researcher checked if the data on each columns were normally distributed using distribution plot. None of the columns are normally distributed.

Several graphs were also used to check the trends and the correlation each columns.
- Distribution plot to determine the distribution of the data in each column
- Line graphs to see the trends of the housing loans over a specific time period
- Line Graph and Scatter Plot to see the trends in housing prices as well as the correlation of prices to housing construction cost index
- Heatmap for checking the correlation of each column in a graphical format

```python
# Check if the the data is normally distributed
# Loop through each column in the DataFrame
for column in housing_df.columns:
    # Create a distribution plot for numeric columns
    if pd.api.types.is_numeric_dtype(housing_df[column]):
        sns.displot(data=housing_df, x=column, kde=True)
        plt.xlabel('Values')
        plt.ylabel('Frequency')
        plt.title(f'Distribution Plot of {column}')
        plt.show();
        
        sns.boxplot(data=housing_df, x=column)
        plt.xlabel('Values')
        plt.ylabel('Frequency')
        plt.title(f'Box Plot of {column}')
        plt.show();
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x="Year", y="Number of House Loans Approved - New Houses", data=housing_df, label="Number of House Loans Approved - New Houses")
sns.lineplot(x="Year", y="Number of House Loans Paid - New Houses", data=housing_df, label="Number of House Loans Paid - New Houses")
sns.lineplot(x="Year", y="Number of House Loans Approved - Second hand Houses", data=housing_df, label="Number of House Loans Approved - Second hand Houses")
sns.lineplot(x="Year", y="Number of House Loans Paid - Second hand Houses", data=housing_df, label="Number of House Loans Paid - Second hand Houses")
# Add labels and title
plt.xlabel('Year')
plt.ylabel('Application')
plt.title('Loans Approved and Paid from 1975 to 2016')
# Show legend
plt.legend()

# Show the line plot
plt.show()
```
